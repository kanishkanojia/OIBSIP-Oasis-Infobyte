## Import the Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly .express as px
```

## Loading the dataset

```python
data = pd.read_csv('/content/Unemployment in India.csv')
data
```

```python
# checking dataset information
data.info()
```

```python
# describing the dataset
data.describe()
```

```python
# check null/missing values
data.isnull().sum()
```

```python
# rename columns
data.columns = ['Region','Date','Frequency','Estimated Unemployment Rate',
                'Estimated Employed','Estimated Labour Participation Rate',
                'Area']
```

```python
# analysing top rows of dataset
data.head()
```

## CHECKING THE CORRELATION BETWEEN THE FEATURE OF DATASET

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style of the plot to 'seaborn-whitegrid'
plt.style.use('seaborn-whitegrid')

# Create a new figure with size 8x6 inches
plt.figure(figsize=(8, 6))

# Compute the correlation matrix using the `corr()` method on the DataFrame `data`
correlation_matrix = data.corr()

# Compute the correlation matrix and plot the heatmap
ax = sns.heatmap(correlation_matrix, annot=True, linewidths=0.5, cmap='coolwarm', center=0)

# Set tick parameters
ax.tick_params(size=8, color='black', labelsize=8, labelcolor='black')

# Show the plot
plt.show()
```

## Estimated no of employee according to different region of india

```python
# plotting histplot

data.columns=['Region','Date','Frequency','Estimated Unemployment Rate',
                'Estimated Employed','Estimated Labour Participation Rate',
                'Area']
plt.title('Indian Unemployment')
sns.histplot(x='Estimated Employed',hue='Region',data=data)
plt.show()
```

## Unemployment rate according to different regions of india

```python
# plotting histplot

plt.figure(figsize=(10,8))
plt.title("Indian Unemployment")
sns.histplot(x="Estimated Unemployment Rate",hue='Region',data=data)
plt.show()
```

## Dashboard to analyze the unemployment rate of each Indian state

```python
import pandas as pd
import plotly.express as px


# Group the data by 'Region', 'Date', and 'Area' columns and calculate the sum of 'Frequency' column
grouped_df = data.groupby(['Region', 'Date', 'Area']).sum().reset_index()

# Create the sunburst chart
figure = px.sunburst(grouped_df, path=['Region', 'Area'],
                     values='Estimated Unemployment Rate',
                     title='Unemployment Rate by State',
                     color_continuous_scale='RdYlGn')

figure.update_layout(width=800, height=800)
figure.show()

```
